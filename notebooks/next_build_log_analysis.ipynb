# Next.js Build Log Analysis

This notebook analyzes the provided Next.js build output, extracts key build events, timing metrics, rendering modes, and the detected vulnerability warning.

```python
build_log = '''○  (Static)   prerendered as static content
ƒ  (Dynamic)  server-rendered on demand
Traced Next.js server files in: 199.779ms
Created all serverless functions in: 228.87ms
Collected static files (public/, static/, .next/static): 42.824ms
Build Completed in /vercel/output [2m]
Vulnerable version of Next.js detected, please update immediately.
'''
print(build_log)
```

## Parse Next.js Build Logs
Load and parse the Next.js build log output to extract key events and status messages.

```python
import re

lines = build_log.strip().splitlines()
parsed = []
for line in lines:
    if line.startswith('○') or line.startswith('ƒ'):
        parsed.append(('render_mode', line.strip()))
    elif 'Traced Next.js server files in:' in line:
        parsed.append(('trace_time', line.split(':',1)[1].strip()))
    elif 'Created all serverless functions in:' in line:
        parsed.append(('serverless_time', line.split(':',1)[1].strip()))
    elif 'Collected static files' in line:
        parsed.append(('static_files_time', line.split(':',1)[1].strip()))
    elif 'Build Completed' in line:
        parsed.append(('build_finished', line.split('Build Completed in',1)[1].strip()))
    elif 'Vulnerable version of Next.js detected' in line:
        parsed.append(('vulnerability_warning', line.strip()))

parsed
```

## Identify Static and Dynamic Rendering Modes
Inspect the build log for static prerendered pages and server-rendered dynamic pages, and explain their differences.

```python
render_modes = [line for line in lines if line.startswith('○') or line.startswith('ƒ')]
render_modes
```
